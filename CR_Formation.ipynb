### Troisième projet : Boston Housing  

#### Introduction :  

On commence à entrer dans des projets plus intéressant, mais plus complexes. Celui-ci s'appuie sur un jeu de donnée à propos de l'immobilier dans des quartiers de Boston.
Il s'agit d'estimer le prix d'un bien en se basant sur un certain nombre de paramètres (nombre de pièces, taux de criminalité, distance au centre-ville, etc.)  

#### But : 

Le projet Boston Housing se concentre cette fois sur la régression (multi-)linéaire. Il ne s'agit plus d'implémenter un modèle, mais de le choisir, l'évaluer et le corriger.

#### Déroulé :

Le déroulé de chaque projet se ressemble beaucoup pour l'instant, et tant mieux. On commence par observer le jeu de donnée. Ici on remarque qu'il n'y a aucun "NaN" ou valeur manquante dans les colonnes. Pas de remplissage à faire.

Il s'agissait ensuite d'observer les corrélations entre l'output (prix du bien) et les caractéristiques, pour avoir une première idée de celles impactant le plus le prix.

Comme auparavant, on prépare le jeu pour l'entrainer. On sépare l'output (prix) des inputs (caractéristiques). On standardise les données (centrer-réduire) pour accorder le même poids à chaque variable.

##### Régression linéaire : 

Le modèle appliquer dans ce projet et la régression linéaire. Celui-ci s'occupe d'exprimer, de manière linéaire, l'output en fonction des inputs, en trouvant les coefficients optimaux pour chaque variable tel que : 

prix($[var_1, var_2, ..., var_n]$) = $c_1 \times var_1 + ... + c_n \times var_n$ 
où : 
- $var_i$ pour i = 1,...n sont les variables qui jouent dans le prix de bien
- $c_i$ pour i = 1,...,n sont les coefs qui définissent l'impact de chaque variable dans 
           le prix final.

##### Evaluation du modèle : 

Après avoir appliquer le modèle sur notre jeu de donnée, il est pertinent d'examiner certaines mesures : 
- **MAE** : Erreur moyenne absolue
  Une mesure très intéressante puisqu'elle est facilement interprétable : 
Elle calcule la distance moyenne entre la prédiction et la valeur réelle du bien. Ainsi pour MAE = 3.19, on conclura que le modèle prédira le prix avec une erreur moyenne de $\pm 3190 \ {$} $.
- **RMSE** : Erreur quadratique moyenne
  Cette mesure est similaire au MAE, mais pénalise davantage les grosses erreurs de prédiction (car élevées au carré).
- $R^2$ : proportion de variance expliquée. (coefficient de détermination)
  Entre 0 et 1
  -plus c'est proche de 0, moins le modèle est pertinent et fiable
  - plus c'est proche de 1, plus le modèle est fiable.
  - 0.75 signifie que 75% de la variance


On peut ensuite comparer les valeurs prédites (abscisse) avec les valeurs réelles (ordonnée) sur un graphe pour observer la linéarité du modèle. Plus les points sont sur y = x, plus le modèle est bon.

### $1^{ère}$ formation - Simple Linear Regression   

Cette première formation est prévue pour repartir de 0, surtout pour m'assurer de mes fondations, et gagner confiance en mes compétences.

 #### Quelques compétences clés : 

##### **matplotlib**

> scatter, plot, show, hist (déja maitrisé en Licence)

##### **statsmodels.api**

> **OLS.from_formula()** : modèle de régression linéaire de la bibliothèque citée  
cette méthode sert à préparer le modèle. Elle réunit les éléments nécessaire au fonctionnement du modèle.

> **fit()** : méthode appliquée au modèle qui va le lancer, donner les coefficients de la formule linéaire ainsi que plusieurs méthodes importantes :
> - fitted values (valeurs prédites par le modèle)
> - résidus (différence entre valeurs réelles et valeurs prédites)
> - params : les coefficients de la formule linéarisée

##### **Distribution normale/ Hypothèse de normalité**

>C'est un outil que je n'avais pas vu en licence. Il sert à vérifier si les résidus sont dispersés de manière normale (gaussienne).  
>Le cas échéant, on conclura que le modèle linéaire n'est pas une bonne solution de prédiction. 

##### **Homoscedasticity Assumption (homoscédasticité)**

>Deuxième outil que je n'avais pas vu. Il s'agit de générer un graphe qui met en relation les valeurs prédites avec les résidus.  
>Le but ici est d'obtenir une distribution uniforme, bien "étalée".

##### **caractéristiques qualitatives (categorical)**

>Ici, il s'agissait de traiter des données binaires (0/1)  
>On implémente le modèle de la même manière
